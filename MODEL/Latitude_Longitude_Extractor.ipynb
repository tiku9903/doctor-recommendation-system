```python
import pandas as pd
import numpy as np
import requests
def get_lat_lng(address):
    api_key = ''
    url = f'https://geocode.maps.co/search?q={address}&api_key={api_key}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        if data:
            latitude = data[0]['lat']
            longitude = data[0]['lon']
            return latitude, longitude
        else:
            return np.nan, np.nan  
    except Exception as e:
        print(f"An error occurred: {e}")
        return np.nan, np.nan  

df_subset = df.iloc[:90000].copy()  

latitudes = []
longitudes = []
for _, row in df_subset.iterrows():
    address_parts = [str(row['adr_ln_1']), str(row['City/Town']), str(row['State'])]
    address = ', '.join(address_parts)
    try:
        latitude, longitude = get_lat_lng(address)
    except Exception as e:
        print(f"Error occurred for address {address}: {e}")
        latitude, longitude = np.nan, np.nan
    latitudes.append(latitude)
    longitudes.append(longitude)

df_subset['Latitude'] = latitudes
df_subset['Longitude'] = longitudes
```

```python
df_subset2 = df.iloc[:38528].copy()  # Create a copy of the DataFrame
df_subset2['Latitude'] = latitudes
df_subset2['Longitude'] = longitudes

# Save to a new CSV file
df_subset2.to_csv('addresses_with_lat_lng1.csv', index=False)
```
